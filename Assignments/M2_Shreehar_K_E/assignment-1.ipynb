## 9. Generate the series of dates from 1st Jan, 2023 to 10th Feb, 2023

```python
import datetime
start = datetime.datetime.strptime("01-01-2023","%d-%m-%Y")
end = datetime.datetime.strptime("10-02-2023","%d-%m-%Y")
a = pd.date_range(start, end)
pd.Series(data = a)
```
